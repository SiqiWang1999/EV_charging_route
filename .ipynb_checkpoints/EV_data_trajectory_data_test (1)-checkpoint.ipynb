```python
import pandas as pd
import matplotlib.pyplot as plt
import descartes
import geopandas as gpd
from shapely.geometry import Point, Polygon

%matplotlib inline
```

```python
from pathlib import Path

print(Path.cwd())
```

```python
df = pd.read_csv('/Users/wangsiqi/Downloads/VED_DynamicData_Part1/VED_171101_week.csv')
crs ={'init': 'epsg:4326'}
df.head()
```

```python
list_of_column_names = list(df.columns)
print('List of column names : ',list_of_column_names)
```

```python
geometry = [Point(xy) for xy in zip( df['Longitude[deg]'], df['Latitude[deg]'])]
geometry[:3]
```

```python
geo_df = gpd.GeoDataFrame(df, crs=crs, geometry = geometry)
geo_df.head()
```

```python
fig, ax = plt.subplots(figsize = (15,15))
geo_df.geometry.plot(ax = ax, color = 'blue', marker = "^" )

plt.legend(prop={'size':15})
```
